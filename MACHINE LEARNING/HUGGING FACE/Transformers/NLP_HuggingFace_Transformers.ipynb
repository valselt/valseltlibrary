### 1. Zero-Shot Classifier

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

output = classifier(
    "This is a course about the Transformers library. This course covers the popular Transformers library used for text analysis",
    candidate_labels=["education", "technology", "natural language processing", "machine learning"],
)
print("Classification Results:")
for label, score in zip(output['labels'], output['scores']):
    print(f"Label: {label} | Score: {score:.4f}")
```

Zero-shot classification adalah metode dalam pemrosesan bahasa alami (NLP) yang memungkinkan model untuk mengklasifikasikan teks ke dalam kategori yang belum pernah dilatih sebelumnya.

> Pada kode diatas, zero-shot classification dijalankan menggunakan model `facebook/bart-large-mnli`.





Teks `"This is a course about the Transformers library. This course covers the popular Transformers library used for text analysis"` akan diklasifikasikan kedalam beberapa daftar label kandidat yang telah ditentukan yaitu `"education", "technology", "natural language processing", "machine learning"`.

Hasil menunjukkan bahwa teks tersebut mendapat skor confidence yang dinilai dari rentang 0-1 sebagai berikut :

1. technology: 0.4868
2. education: 0.3116
3. natural language processing: 0.1433
4. machine learning: 0.0582

Dari skor confidence diatas, dapat diketahui bahwa model yakin bahwa teks yang diberikan lebih masuk kedalam kategori `technology` daripada yang lainnya.

### 2. Text Generator

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

output = generator(
    "In this Hugging Face's Transformers library course, we will teach you how to"
)

print("Generated Text:")
for i, generated in enumerate(output):
    print(f"Output {i+1}: {generated['generated_text']}")
```

text-generation (pembuatan teks) adalah proses dalam NLP untuk menghasilkan teks baru berdasarkan input atau konteks yang diberikan.

> Pada kode diatas, text-generation dijalankan menggunakan model `gpt2`.



Dengan inputan seperti ini :

> In this Hugging Face's Transformers library course, we will teach you how to

Hasil yang ditunjukkan adalah seperti ini :
 > "In this Hugging Face's Transformers library course, we will teach you how to design a robot that can make up to 1 Billion cubic feet of water, using only one of the available materials. You will learn how to construct the robot by creating all"

Dalam hal ini, model belum secara sempurna mengerti konteks yang diberikan oleh input dan menghasilkan output yang berbeda dari pemikiran orang yang menginput.

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

output = generator(
    "I have a dream about getting a car that",
    max_length=50,
    num_return_sequences=2,
)

print("Generated Text:")
for i, generated in enumerate(output):
    print(f"Output {i+1}: {generated['generated_text']}")
```

Contoh diatas adalah penerapan lain untuk menjalankan `generator`.

Bagian `max_length` ditujukan untuk mengatur panjang maksimum teks yang dihasilkan.

> Pada Contoh diatas `max_length=50` artinya panjang maksimum kata yang keluar adalah 50 token atau kata.




Bagian `num_return_sequences` ditujukan untuk meminta model untuk menghasilkan dua hasil dari generator.

> Pada Contoh diatas `num_return_sequences=2` artinya meminta model untuk mengeluarkan 2 output teks.

### 3. Fill-Mask

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
input_sentence = "I want to make me a beautiful <mask> to feeling good about myself."
output = unmasker(input_sentence, top_k=2)

print("Predicted Words for the Mask:")
for i, prediction in enumerate(output):
    predicted_word = prediction['token_str']
    score = prediction['score']
    full_sentence = input_sentence.replace("<mask>", predicted_word)

    print(f"Prediction {i+1}: {predicted_word} (Score: {score:.4f})")
    print(f"Full Sentence: {full_sentence}\n")
```

Fill-mask adalah sebuah teknik dalam pemrosesan bahasa alami (NLP) yang digunakan untuk mengisi kata yang hilang dalam kalimat.

Dengan inputan seperti ini :

> "I want to make me a beautiful < mask > to feeling good about myself."

Token `<mask>` adalah kata yang hilang dan perlu diisi oleh model

Hasil yang ditunjukkan adalah seperti ini :
 > 1. I want to make me a beautiful guide to feeling good about myself.

 > 2. I want to make me a beautiful monument to feeling good about myself.

Dari hasil ke 1, model menggunakan kata `guide` untuk mengisi mask dengan skor confidence 0.1227. Sedangkan untuk kata `monument`, model mengisinya dengan skor confidence 0.1146

### 4. Named Entity Recognition (NER)

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
text = "My Name Is Muhammad Ivan Aldorino. I am a student from Universitas Negeri Semarang and now I attended MSIB BATCH 7 at Infinite Learning. I think Semarang is a great place to take a degree on."
output = ner(text)

print("Named Entities:")
for entity in output:
    entity_str = entity['word']
    entity_type = entity['entity_group']
    start = entity['start']
    end = entity['end']

    print(f"Entity: {entity_str} | Type: {entity_type} | Position: ({start}, {end})")
```

Named Entity Recognition (NER) adalah sebuah teknik dalam pemrosesan bahasa alami (NLP) yang digunakan untuk mengidentifikasi dan mengklasifikasikan entity bernama dalam teks.

`grouped_entities=True` menunjukkan bahwa hasil pengenalan entities akan dikelompokkan untuk mempermudah membacanya.



Hasil dari identifikasi oleh model adalah :


> 1. Entity Pertama</br>
- entity_group: PER (Person)
- score: 0.998613
- word: "Muhammad Ivan Aldorino"
- start: 11
- end: 33</br></br>
artinya, model mengenali kata "Muhammad Ivan Aldorino" sebagai nama orang dan memasukkannya kedalam entity group `PER` dengan skor confidence 0.9986

> 2. Entity Kedua</br>
- entity_group: ORG (Organization)
- score: 0.9639477
- word: "Universitas Negeri Semarang"
- start: 56
- end: 83</br></br>
artinya, model mengenali kata "Universitas Negeri Semarang" sebagai nama organisasi dan memasukkannya kedalam entity group `ORG` dengan skor confidence 0.9639


> 3. Entity Ketiga</br>
- entity_group: ORG (Organization)
- score: 0.95237684
- word: "Infinite Learning"
- start: 120
- end: 137</br></br>
artinya, model mengenali kata "Infinite Learning" sebagai nama organisasi dan memasukkannya kedalam entity group `ORG` dengan skor confidence 0.9523

> 4. Entity Keempat</br>
- entity_group: LOC (Location)
- score: 0.9933
- word: "Semarang"
- start: 147
- end: 155</br></br>
artinya, model mengenali kata "semarang" sebagai nama organisasi dan memasukkannya kedalam entity group `LOC` dengan skor confidence 0.9933

### 5. Question Answering

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "Semarang terkenal akan Bandeng Presto sehingga sering dijadikan opsi oleh-oleh bagi para turis."
question = "Kota mana yang terkenal dengan Bandeng Presto?"

result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

Question-Answering (QA) adalah teknik dalam NLP yang fokus pada kemampuan suatu sistem untuk menjawab pertanyaan berdasarkan informasi yang diberikan.



Dalam question-answering, dibutuhkan sebuah konteks agar model dapat menjawab pertanyaan yang diberikan.

`context` adalah sebuah konteks yang diberikan kepada NLP terkait informasi tertentu. Setelah diberikan konteks, maka pertanyaan dapat dimasukkan pada `question` yang sesuai dengan konteksnya.


Sebagai contoh, akan dimasukkan `context` seperti ini : </br>
"Semarang terkenal akan Bandeng Presto sehingga sering dijadikan opsi oleh-oleh bagi para turis."
</br></br>
lalu akan diberikan `question` seperti ini :</br>
Kota mana yang terkenal dengan Bandeng Presto?


Setelah code dijalankan, akan keluar `answer` dari sebuah `question` yang dianalisa menggunakan `context` yang ada.

Jawaban yang keluar adalah "Semarang"

#### 5.1 Contoh Lainnya

### 6. Sentiment Analysis

```python
from transformers import pipeline


classifier = pipeline("sentiment-analysis")


text = "Nasi padang and fried chicken are the perfect combination, and when someone buys it for me, I feel so delighted!"
output = classifier(text)

print("Sentiment Analysis Result:")
for result in output:
    label = result['label']
    score = result['score']
    print(f"Sentiment: {label} | Score: {score:.4f}")
```

Sentiment Analysis merupakan salah satu teknik dalam NLP untuk mengidentifikasi dan mengklasifikasikan emosi atau opini yang terkandung dalam teks.


Sebagai Contoh, saya menuliskan kata "Nasi padang and fried chicken are the perfect combination, and when someone buys it for me, I feel so delighted!".

Teks tersebut kemudian diterima oleh model dan dilakukan analisa sentimen untuk menentukan apakah harus dilabeli POSITIVE atau NEGATIVE.



Sebagai contoh, kata diatas diklasifikasi oleh model sebagai kata yang `POSITIVE` dan memiliki skor confidence yaitu 0.9997

```python
from transformers import pipeline


classifier = pipeline("sentiment-analysis")


text = "Matcha and Chocolate are the worst drink combination ever existed. I wonder how people can drink such a thing!"
output = classifier(text)

print("Sentiment Analysis Result:")
for result in output:
    label = result['label']
    score = result['score']
    print(f"Sentiment: {label} | Score: {score:.4f}")
```

Diatas merupakan contoh ketika model menganalisis sentimen yang ada didalam kata dan melabelinya menjadi `NEGATIVE` dengan skor confidence yaitu 0.9998
